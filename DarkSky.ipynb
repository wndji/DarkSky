```python
df = pd.DataFrame([])
unix_day = 24*60*60
time = 1514782800
u = 'https://api.darksky.net/forecast/d1cf5ff15c2f2d13c6519d65bfaf5fed/40.77898,-73.96925,'
x = '?exclude=currently,alerts,hourly,flags,minutely'

for i in range(0,365):
    url = u + str(time) + x
    r = requests.get(url)
    data = r.json()
    data_output = pd.DataFrame.from_dict(json_normalize(data['daily']['data']))
    #append to df
    df = df.append(data_output)
    time = time+unix_day
```
